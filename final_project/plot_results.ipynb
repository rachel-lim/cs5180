```python
import os
import glob

import matplotlib.pyplot as plt
import numpy as np

from utils.plotting import plot_curves, moving_average
```

```python
from typing import List
def plot_curves(arr_list: List[np.array], legend_list: List[str], color_list: List[str], ylabel: str, title: str = None, xlabel: str = "Episodes"):
    """
    Args:
        arr_list: list of results arrays to plot
        legend_list: list of legends corresponding to each result array
        color_list: list of color corresponding to each result array
        ylabel: label for the y axis
        title: plot title
        xlabel: label for the x axis
    """
    # set the figure type
    plt.clf()
    fig, ax = plt.subplots()

    ax.set_ylabel(ylabel)
    ax.set_xlabel(xlabel)

    # ploth results
    h_list = []
    for arr, legend, color in zip(arr_list, legend_list, color_list):
        # compute the standard error
        arr_err = arr.std(axis=0) / np.sqrt(arr.shape[0])
        # plot the mean
        h, = ax.plot(range(0, 1000*arr.shape[1], 1000), arr.mean(axis=0), color=color, alpha = 0.3, label=legend)
        ax.plot(range(0, 1000*arr.shape[1], 1000), moving_average(arr.mean(axis=0), window_size=20), color=color, label=f"{legend} moving average")
        # plot the confidence band
        arr_err = 1.96 * arr_err
        ax.fill_between(range(arr.shape[1]), arr.mean(axis=0) - arr_err, arr.mean(axis=0) + arr_err, alpha=0.3,
                        color=color)
        # save the plot handle
        h_list.append(h)

    # plot legends
    # ax.legend(handles=h_list)
    ax.set_ylim([-23, 11])
    ax.grid()
    # ax.legend(loc="upper left", bbox_to_anchor=(0, -.1))
    ax.legend()
    ax.set_title(title)
    plt.show()
```

```python
eval_regular_reg_maze = []
eval_regular_rot_maze = []
returns_regular = []
for file in glob.glob(os.path.join("results", "maze_regular*")):
    with open(file, "rb") as f:
        eval_regular_reg_maze.append(np.load(f))
        eval_regular_rot_maze.append(np.load(f))
        returns_regular.append(np.load(f))
eval_equi_reg_maze = []
eval_equi_rot_maze = []
returns_equi = []
for file in glob.glob(os.path.join("results", "maze_equi*")):
    with open(file, "rb") as f:
        eval_equi_reg_maze.append(np.load(f))
        eval_equi_rot_maze.append(np.load(f))
        returns_equi.append(np.load(f))

plot_curves([np.array(eval_regular_reg_maze), np.array(eval_equi_reg_maze)], 
            ["Regular CNN", "Equivariant CNN"], ["C0", "C1"], "Returns", "Regular Maze - Evaluated Returns")
plot_curves([np.array(eval_regular_rot_maze), np.array(eval_equi_rot_maze)], 
            ["Regular CNN", "Equivariant CNN"], ["C0", "C1"], "Returns", "Rotated Maze - Evaluated Returns")
```
